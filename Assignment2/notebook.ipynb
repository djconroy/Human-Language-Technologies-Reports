```python
import string
```

```python
def get_lines(file_name):
    with open(file_name, encoding='utf8') as file:
        return file.readlines()
```

```python
lines = get_lines('100west.txt')
```

```python
lines = list(map(lambda line: ''.join(c for c in line.strip().lower() if c not in string.punctuation) + '\n', lines))
```

```python
with open('100west_for_lm.txt', 'w') as file:
    file.write(''.join(lines))
```

```python
def get_file(file_name):
    with open(file_name, encoding='utf8') as file:
        return file.read()
```

```python
lm = get_file('1198.lm')
```
